```python
import pandas as pd

df = pd.read_csv("Placement_Data_Full_Class.csv")
df.head()
```

```python
df.columns
```

```python
# Check missing values in each column
df.isnull().sum()
```

```python
# Fill missing salary values (students not placed → no salary)
df['salary'] = df['salary'].fillna(0)
```

```python
# Remove duplicate records (if any)
df.drop_duplicates(inplace=True)
```

```python
# Clean text columns (remove spaces and standardize format)
df['gender'] = df['gender'].str.strip().str.upper()
df['workex'] = df['workex'].str.strip().str.upper()
df['status'] = df['status'].str.strip().str.capitalize()
df['specialisation'] = df['specialisation'].str.strip()
```

```python
# Convert salary column to integer type
df['salary'] = df['salary'].astype(int)
```

```python
# Create a new column to categorize salary levels
df['salary_category'] = df['salary'].apply(lambda x:
    'No Salary' if x == 0 else
    'Low' if x < 300000 else
    'Medium' if x < 600000 else
    'High'
)
```

```python
# Save cleaned dataset to CSV file
df.to_csv("placement_cleaned_dataset.csv", index=False)
```
